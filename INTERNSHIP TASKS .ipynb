```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import pickle
```

```python
train_df=pd.read_excel(r'C:\Users\Hi\Downloads\train.xlsx')
train_df
```

```python
train_df.isnull().sum()
```

```python
train_df.info()
```

```python
# Selecting numerical columns for clustering
numerical_cols=train_df.select_dtypes(include=['float64','int64']).columns
a= train_df[numerical_cols]
a
```

```python
# Standardizing the data
scaler= StandardScaler()
a_scaled= scaler.fit_transform(a)
a_scaled
```

```python
# Perform KMeans clustering
kmeans = KMeans(n_clusters=4, random_state=40)
kmeans.fit(a_scaled)
```

```python
# Assign clusters to the data
train_df['Cluster'] = kmeans.labels_
train_df['Cluster'].value_counts()
```

```python
def predict_cluster(data_point):
    scaled_data_point = scaler.transform(data_point.reshape(1, -1))
    cluster = kmeans.predict(scaled_data_point)
    return cluster[0]

# Example: Predict cluster for the first data point
data_point = a_scaled[0]
predicted_cluster = predict_cluster(data_point)

train_df.head()
```
